# Processing DataFrames with Pandas

This is the notebook for processing data with Pandas, and compare code + running time with the same process on Pandas (available on `processing_with_pyspark.ipynb`).

```python
import pandas as pd
```

```python
df = pd.read_csv('tweets.csv', header=None, escapechar='\\', encoding='latin-1')
df.head()
```

```python
from datetime import datetime

start = datetime.now()

result = df.groupby(3)[3] \
  .count() \
  .sort_values() \
  .head(10)

print("Consumed time:", datetime.now() - start)
```
